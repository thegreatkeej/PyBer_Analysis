```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="yellow", linewidth=5, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a horizontal plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="red", label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Create a title.
ax.title("PyBer Fare by Month")
# Add the legend.
ax.legend()
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Add the legend.
ax.legend()
# Create a title.
ax.title("PyBer Fare by Month")
```

```python
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]


x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(x_axis, y_axis, color="red", label='Chicago')

plt.xlim(0,50)
plt.xlabel("Fare($)")
plt.ylabel("Date")


plt.title("PyBer Fare by Month")

plt.legend()
plt.gca().invert_yaxis()
```

```python
# For loop style bubble chart
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List comprehension style bubble chart
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Object-oriented List comprehension style bubble chart
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
