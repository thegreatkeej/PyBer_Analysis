### 1.06 Instructor Do: Merging DataFrames (0:15)

````python
# Dependencies
import pandas as pd
````

````python
# Read in the following data into a DataFrame.
raw_data_info = {
    "customer_id": [112, 403, 999, 543, 123],
    "name": ["John", "Kelly", "Sam", "April", "Bobbo"],
    "email": ["jman@gmail", "kelly@aol.com", "sports@school.edu", "April@yahoo.com", "HeyImBobbo@msn.com"]
}
info_df = pd.DataFrame(raw_data_info, columns=["customer_id", "name", "email"])
info_df
````

````python
# Read in the following data into a DataFrame.
raw_data_items = {
    "customer_id": [403, 112, 543, 999, 654],
    "item": ["soda", "chips", "TV", "Laptop", "Cooler"],
    "cost": [3.00, 4.50, 600, 900, 150]
}
items_df = pd.DataFrame(raw_data_items, columns=[
                        "customer_id", "item", "cost"])
items_df
````

````python
# Merge two dataframes using an inner join (default).
merge_df = pd.merge(info_df, items_df, on="customer_id")
merge_df
````

````python
# Merge two dataframes using an outer join
merge_outer_df = pd.merge(info_df, items_df, on="customer_id", how="outer")
merge_outer_df
````

````python
# Merge two dataframes using a left join
merge_left_df = pd.merge(info_df, items_df, on="customer_id", how="left")
merge_left_df
````

````python
# Merge two dataframes using a right join
merge_right_df = pd.merge(info_df, items_df, on="customer_id", how="right")
merge_right_df
````

### 1.07 Instructor Do: Formatting and Mapping (0:05)

````python
pi = 123458527453.141592653589793
````

````python
"{:,.2f}".format(pi)
````

## We can use f-strings to print out numbers as strings, but sometimes the numbers are unwieldy

````python
print(f'The value of pi is approximately: {pi}')
````

## We can also use f-strings to format numbers so that they are easier to read

Inside the curly brackets, we can put a colon after the expression we want to display, and after the colon, we can specify how we want the number displayed.

For float values, we use a decimal point followed by the number of decimal digits to display, and then the letter `f` for `float`

````python
print(f'The value of pi to two digits is ${pi:,.2f}')
````

* We can use a comma separator to make large numbers easier to read

````python
number_of_maniacs = 10000

print(f'{number_of_maniacs:,} Maniacs')
````

* And we can convert values to percentages

````python
print(f'Two-thirds is roughly {2/3:.1%}')
````

* The squiggly bracket formatting is how f-strings got their name! The `f` is for `format`, which is a method that comes for free with all string objects. We can write a format string and then call the `format` method on a value, and it will return the formatted string

````python
"{:.2f}".format(pi)
````

    * `Units Sold` with comma notation
    * `Unit Price`, `Unit Cost`, `Total Revenue`, `Total Cost`, and `Total Profit` with a dollar sign, comma notation, and to two decimal places.

---

````python
import pandas as pd
````

````python
# Mapping lets you format an entire DataFrame
file = "08-Evr_FormatMapping/Solved/Resources/sales_data.csv"
sales_data_df = pd.read_csv(file)
sales_data_df.head()
````

````python
sales_data_df.dtypes
````

````python
sales_data_df.columns
````

````python
sales_data_df2= sales_data_df.copy()
````

````python
# Use map() to format:
# Units Sold with comma notation
# Unit Price, Unit Cost, Total Revenue, Total Cost, and Total Profit 
# with a dollar sign, comma notation, and to two decimal places.

sales_data_df2["Units Sold"] = sales_data_df["Units Sold"].map("{:,.0f}".format)
sales_data_df2["Unit Price"] = sales_data_df["Unit Price"].map("${:,.2f}".format)
sales_data_df2["Unit Cost"] = sales_data_df["Unit Cost"].map("${:,.2f}".format)
sales_data_df2["Total Revenue"] = sales_data_df["Total Revenue"].map("${:,.2f}".format)
sales_data_df2["Total Cost"] = sales_data_df["Total Cost"].map("${:,.2f}".format)
sales_data_df2["Total Profit"] = sales_data_df["Total Profit"].map("${:,.2f}".format)
sales_data_df2.head()
````

````python
# Mapping has changed the datatypes of the columns to strings
sales_data_df2.dtypes
````
